```python
dataset1=dataset.fillna(0)
```

```python
dataset1['Day']=pd.to_datetime(dataset1.Day)
```

```python
dataset1['day']=dataset1["Day"].dt.day
dataset1['month']=dataset1["Day"].dt.month
```

```python
dataset1['month_name']=dataset1['Day'].dt.month_name().str[ :3]
dataset1['weekday']=dataset1['Day'].dt.day_name().str[ :3]
```

```python
dataset1['Date']=dataset1['day'].astype(str)+" "+dataset1['month_name']+" " +"( "+dataset1['weekday']+")"
dataset1.sort_values(by='day',ascending=True,inplace =True)
dataset1.head(2)
```

```python
dataset1.drop([ 'Day', 'Delivery status', 'Delivery level','CTR (all)',
        'Impressions', 'CPM (cost per 1,000 impressions)',
       'Reach', 'Frequency',  'Video plays',
       'Video plays at 100%', 
       'Result Type',  'Purchase ROAS (return on ad spend)',
       'Attribution setting', 'Cost per result', 'Reporting starts',
       'Reporting ends', 'day', 'month', 'month_name', 'weekday'],
  axis=1, inplace=True)
```

```python
dataset1
```

```python
dataset1.columns
dataset1.columns = dataset1.columns.str.replace('Results', 'No Of Leads') 
dataset1.columns = dataset1.columns.str.replace('Amount spent (INR)', 'AD_spent') 
dataset1.columns=dataset1.columns.str.replace('CTR (link click-through rate)','CTR')
```

```python
dataset1.head()
```

```python
dataset1['Cost per lead']=round((dataset1['AD_spent']/dataset1['No Of Leads']),2)
dataset1['Member Passed %']=round((dataset1['Landing page views']/dataset1[ 'Link clicks']*100),2)
dataset1['Lp Conversion %']=round((dataset1[ 'No Of Leads']/dataset1[ 'Landing page views']*100),2)

dataset1.set_index("Date",inplace=True)
```
